3. 중복 수업 확인 (수업 번호 기준으로 이미 유니크한 과목들로 데이터셋을 이루고 있지만, 학수번호가 중복되는 경우가 있음)

```python
n_rows = len(df)

# 수업번호
n_unique_classes = df['수업번호'].nunique()
print(f"총 행: {n_rows}, 수업번호 고유값 개수: {n_unique_classes}")

# 같지 않으면 중복이 있다는 의미
if n_unique_classes != n_rows:
    print("→ 수업번호에 중복이 있습니다.")

# 학수번호도 동일하게
n_unique_codes = df['학수번호'].nunique()
print(f"총 행: {n_rows}, 학수번호 고유값 개수: {n_unique_codes}")
```

4. 평균 텍스트 길이
- 주차별 강의계획 별 텍스트 평균 길이
- 1~16주차 강의계획 합친 텍스트 평균 길이
- 수업 개요 평균 텍스트 길이

```python
# 주차별 강의계획 평균 텍스트 길이
week_cols = [f"{i}주차 강의계획" for i in range(1, 17)]

avg_lengths = {}
for col in week_cols:
    non_null = df[col].dropna().astype(str)
    non_empty = non_null[non_null.str.len() > 0]
    avg_lengths[col] = non_empty.str.len().mean()

avg_lengths = pd.Series(avg_lengths)
print("주차별 강의계획 평균 텍스트 길이 (빈 값 제외):")
print(avg_lengths)
```

```python
# 1-16주차 강의계획 평균 텍스트 길이
week_cols = [f"{i}주차 강의계획" for i in range(1, 17)]

df['combined'] = (
    df[week_cols]
    .fillna('')                
    .astype(str)               
    .agg(' '.join, axis=1)     
)

lengths = df['combined'].str.len()
avg_length = lengths.mean()

print(f"1~16주차 텍스트 길이 평균: {avg_length:.2f}자")
```

```python
overview = df['교과목 개요'].dropna().astype(str)
non_empty = overview[overview.str.len() > 0]      
lengths = non_empty.str.len()                    

avg_length = lengths.mean()

print(f"교과목 개요 평균 텍스트 길이: {avg_length:.2f}자")
```

주차별 강의 계획에서 자주 쓰이는 텍스트 키워드

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(stop_words='english', max_features=50)
bag = vec.fit_transform(df['combined'])
print(pd.Series(vec.get_feature_names_out(), bag.sum(axis=0).A1).sort_values(ascending=False).head(20))
```

```python
# (1) NaN 제거 & 문자열 변환
corpus = df['교과목 개요'].dropna().astype(str)

# (2) 벡터라이저로 상위 50개 단어 사전 구축 및 변환
vec = CountVectorizer(stop_words='english', max_features=50)
bag = vec.fit_transform(corpus)

# (3) 단어별 총 등장 횟수 계산 & 상위 20개 출력
freq = bag.sum(axis=0).A1
terms = vec.get_feature_names_out()
top20 = pd.Series(freq, index=terms).sort_values(ascending=False).head(20)

print(top20)
```
